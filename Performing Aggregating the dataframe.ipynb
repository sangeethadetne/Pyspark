```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.getOrCreate()
```

```python
data_path = "/Users/sangeethadetne/Desktop/BigData/SampleData/"
```

```python
file_path = data_path +"Jan_2019_ontime.csv"
```

```python
df1= spark.read.format("csv").option("header","true").load(file_path)
```

```python
df1.show()
```

```python
df2 = df1.groupBy("DAY_OF_MONTH").count().orderBy("DAY_OF_MONTH").show()
```

```python
df3 = df1.groupBy("DAY_OF_MONTH").agg({"DISTANCE":"max"}).orderBy("DAY_OF_MONTH").show()
```

```python
df4 = df1.groupBy("DAY_OF_MONTH").agg({"DISTANCE":"mean"}).orderBy("DAY_OF_MONTH").show()
```
